```python
import metis
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.io import mmread
```

```python
G=nx.erdos_renyi_graph(n=20,p=0.3,seed=42)
# Read the dendrimer matrix, remove self-loops
# A = mmread("/home/aryan/Downloads/dendrimer_matrix.mtx")
# Graph = nx.convert_node_labels_to_integers(nx.Graph(A))
# Graph.remove_edges_from(nx.selfloop_edges(Graph))
```

```python
k=[2,4,6]
for nparts in k:
    (edgecuts, membership) = metis.part_graph(G, nparts=nparts)
    partition_map = {i: p for i, p in enumerate(membership)}

    # Count cut edges
    cut_edges = sum(1 for u, v in G.edges() if partition_map[u] != partition_map[v])

    print("Cut edges:", cut_edges)
```

```python
#coloring the graph into segments
color_map=[partition_map.get(i,0) for i in range(len(G.nodes))]
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, node_color=color_map, with_labels=True, cmap=plt.cm.Set2, edge_color='gray')
plt.title("Partitioned Graph")
plt.savefig('output-metis.png')
```
